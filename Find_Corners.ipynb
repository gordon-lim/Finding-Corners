```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
# Read image
img = cv2.imread("/content/maze.jpg")
# Make copy
img_cpy = np.copy(img)
# Convert to grayscale
img_gray = cv2.cvtColor(img_cpy, cv2.COLOR_BGR2GRAY)
# Display
plt.imshow(img_gray, cmap="gray")
```

```
# Detect edges
edges = cv2.Canny(img_gray,0,200)
# Display edges
plt.imshow(edges, cmap='gray')
```

```
gray = np.float32(edges)
dst = cv2.cornerHarris(gray,2,3,0.04)
plt.imshow(dst, cmap='gray')
```

```
# Make the spots more obvious
corners = cv2.dilate(dst, None)
```

```
np.max(corners)
```

```
plt.imshow(corners, cmap='gray')
```

```
og_image = np.copy(img)
og_image = cv2.cvtColor(og_image, cv2.COLOR_BGR2RGB)
threshold = 0.4*corners.max()
for i in range(corners.shape[0]):
  for j in range(corners.shape[1]):
    if corners[i][j] > threshold:
      cv2.circle(og_image, center=(j,i), radius=30, color=(0,255,0),  thickness=1)
```
